## EXPLORING WORLD DATA

### Listing all countries / regions in our data

```python
df.location.drop_duplicates(keep='first')
```

### Selecting the 'World' data

```python
World = df[df['location']== 'World']
World
```

```python
### Finding the date of the maximum number of deaths all over the world
```

```python
World[World['new_deaths']==World['new_deaths'].max()]['date']
```

### Creating a summary table for the most recent date all over the world

```python
date = World.groupby('date').sum().sort_values(by='date',ascending=False)
date.head(1)
```

```python
World.columns
```

### Calculating the percentage of confirmed cases regarding the world population

```python
World
```

```python
percentage = World.total_cases.sum()/ World.population.sum()
print((percentage ) ,'%')
print(f'which is equivalent to {round( percentage*100)}% ')
```

## VISUALLIZING WORLD DATA

### Confirmed cases (Total Cases) all over the world

```python
Confirmed_cases = World.total_cases
Confirmed_cases
```

```python
plt.plot(World['date'] , World['total_cases'])
plt.xlabel('Days')
plt.ylabel('Confirmed Cases')
plt.title('Evolution of Confirmed Covid-19 Cases over time in the world')
plt.show()
```

### Total deaths cases evolution over time

```python
plt.plot(World.date, World.total_deaths)
plt.xlabel('Days')
plt.ylabel('Number of Deaths')
plt.title('Evolution of Confirmed Covid-19 Cases over time in the world')
plt.show()
```

### New cases all over the world

```python
# Using line-plot
plt.plot(World.date, World.new_cases)
plt.xlabel('Days')
plt.ylabel('Number of new cases')
plt.title('Evolution of Confirmed Covid-19 Cases over time in the world')
plt.show()
```

### Putting it all together

```python
plt.plot(World['date'] , World['total_cases'],label='Confirmed',color='k')
plt.plot(World['date'] , World['total_deaths'],label='Total Death',color='r')
plt.plot(World['date'] , World['new_cases'],label='New',color='b')
plt.plot(World['date'] , World['new_deaths'],label='New Deaths',color='g')
plt.legend(loc=0)
plt.figure(figsize=(200 ,200))
plt.show()
```
